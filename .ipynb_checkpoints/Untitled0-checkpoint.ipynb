```python
data_file='./sample_data/Python_EDP_I/matricule.xlsx'
```

```python
#nombre de ligne par province
```

```python
df_grouped_by_Provencia = df_matricules.groupby(by='Provincia').count().reset_index()
```

```python
df_grouped_by_Provencia
```

```python
#Representation histograme avec seaborn
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize= (30, 10))
sns.barplot(x='Provincia', y='Modelo', data=df_grouped_by_Provencia)
```
